## Matplotlib Example Workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [11,22,33,44]

fig , ax = plt.subplots(figsize=(10,10)) #width,height

ax.plot(x,y)

ax.set(title  = "Simple Plot",
       xlabel = "X-Axis",
       ylabel = "Y-Axis ")
fig.savefig("sample-plot.png")
```

## Subplotes

```python
# Method 1

fig , ((ax , bx),(cx, dx)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
ax.plot(x,x/2);
bx.scatter(np.random.random(10),np.random.random(10));
cx.bar(nut_butter_prices.keys(),nut_butter_prices.values());
dx.hist(np.random.randn(1000));
```

```python
# Method 2

fig , ax = plt.subplots(nrows=2,
                        ncols=2,
                        figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease.head()
```

```python
heart_disease["age"].plot(kind="hist");
```

```python
heart_disease.plot.hist(figsize=(10,30), subplots = True);
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
over_50['age'].unique()
```

```python
over_50.plot(
    kind = "scatter",
    x = "age",
    y = "chol",
    c = "target"
);
```

```python
# OO method

fig , ax = plt.subplots(figsize=(10,6))
over_50.plot(
    kind = "scatter",
    x = "age",
    y = "chol",
    c = "target",
    ax = ax
);
```

```python
# OO method from scratch
fig , ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(
    x = over_50['age'],
    y = over_50['chol'],
    c = over_50['target']
)

ax.set(title = "Heart Disease vs Chelestrol level",
       xlabel = "Age",
       ylabel = "Chelestrol")


ax.legend(*scatter.legend_elements(), title = "Target");

ax.axhline(over_50['chol'].mean(),
           linestyle="--");
```

```python
## Mutiple subplotes

fig , (ax , bx) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize = (10,10),
                               sharex= True);
scatter = ax.scatter(
    x = over_50['age'],
    y = over_50['chol'],
    c = over_50['target']
)

ax.set(title = "Heart Disease vs Chelestrol level",
       ylabel = "Chelestrol")

ax.legend(*scatter.legend_elements(), title = "Target");

ax.axhline(over_50['chol'].mean(),
           linestyle="--");

# Plot 2
scatter = bx.scatter(
    x = over_50['age'],
    y = over_50['thalach'],
    c = over_50['target']
)

bx.set(title = "Heart Disease vs Thalach (Max Heart Rate) level",
       xlabel = "Age",
       ylabel = "Thalach")

bx.legend(*scatter.legend_elements(), title = "Target");

bx.axhline(over_50['thalach'].mean(),
           linestyle="--");

fig.suptitle("Heart Disease Analysis" , fontsize = 16, fontweight ="bold");
```

```python
# OO method from scratch
plt.style.use("seaborn-whitegrid");

fig , ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(
    x = over_50['age'],
    y = over_50['chol'],
    c = over_50['target'],
    cmap = "winter", # Colour
)

ax.set(title = "Heart Disease vs Chelestrol level",
       xlabel = "Age",
       ylabel = "Chelestrol")


ax.legend(*scatter.legend_elements(), title = "Target");

ax.axhline(over_50['chol'].mean(),
           linestyle="--");
```

```python
## Mutiple subplotes

fig , (ax , bx) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize = (10,10),
                               sharex= True);
scatter = ax.scatter(
    x = over_50['age'],
    y = over_50['chol'],
    c = over_50['target'],
    cmap = "winter",
)

ax.set(title = "Heart Disease vs Chelestrol level",
       ylabel = "Chelestrol")
ax.set_xlim(50,80)

ax.legend(*scatter.legend_elements(), title = "Target");

ax.axhline(over_50['chol'].mean(),
           linestyle="--");

# Plot 2
scatter = bx.scatter(
    x = over_50['age'],
    y = over_50['thalach'],
    c = over_50['target'],
    cmap = "winter",

)

bx.set(title = "Heart Disease vs Thalach (Max Heart Rate) level",
       xlabel = "Age",
       ylabel = "Thalach")
bx.set_xlim(50,80)
bx.set_ylim(60,200)
bx.legend(*scatter.legend_elements(), title = "Target");

bx.axhline(over_50['thalach'].mean(),
           linestyle="--");

fig.suptitle("Heart Disease Analysis" , fontsize = 16, fontweight ="bold");
```
